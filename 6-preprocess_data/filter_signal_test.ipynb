```python
import pandas as pd
from pathlib import Path
import shutil
import re

# ==== 경로 설정 ====
SRC_DIR = Path("/nas/research/03-Neural_decoding/4-image/beta")
DST_DIR = Path("/nas/research/03-Neural_decoding/4-image/coco_beta_img")
CSV_PATH = Path("/nas/research/03-Neural_decoding/3-bids/2-derivatives/1-beta/beta_nsd/nsd_stim_info_merged.csv")

# ==== 옵션 ====
DRY_RUN = False   # True면 아무 파일도 쓰지 않고 매핑만 출력
OVERWRITE = False # 같은 이름이 이미 있으면 덮어쓸지 여부

# ==== 유틸 ====
pat = re.compile(r"^coco2017_(\d+)\.jpg$", re.IGNORECASE)

def safe_int(x):
    try:
        return int(x)
    except Exception:
        return None

# ==== CSV 로드 & 매핑 준비 ====
df = pd.read_csv(CSV_PATH)
# 혹시 공백/대소문자 문제 방지
df.columns = df.columns.str.strip()

if not {"num", "cocoId"}.issubset(set(df.columns)):
    raise ValueError(f"CSV에 'num', 'cocoId' 컬럼이 있어야 합니다. 현재 컬럼: {list(df.columns)}")

# 정수 변환(문자/float로 저장된 경우 대비)
df["num"] = pd.to_numeric(df["num"], errors="coerce").astype("Int64")
df["cocoId"] = pd.to_numeric(df["cocoId"], errors="coerce").astype("Int64")

# NaN 있는 행 제거(안전)
map_df = df.dropna(subset=["num", "cocoId"]).copy()
num_to_coco = dict(zip(map_df["num"].astype(int), map_df["cocoId"].astype(int)))

# ==== 출력 폴더 생성 ====
DST_DIR.mkdir(parents=True, exist_ok=True)

# ==== 처리 ====
stats = {
    "total_found": 0,
    "renamed": 0,
    "skipped_bad_name": 0,
    "skipped_no_mapping": 0,
    "collision": 0,
}

for src_path in sorted(SRC_DIR.glob("coco2017_*.jpg")):
    stats["total_found"] += 1
    m = pat.match(src_path.name)
    if not m:
        stats["skipped_bad_name"] += 1
        print(f"[SKIP:NAME] 패턴 불일치: {src_path.name}")
        continue

    original_num = int(m.group(1))
    lookup_num = original_num - 1  # <-- 요구사항: 파일명 {number}-1 를 num에서 찾음

    coco_id = num_to_coco.get(lookup_num, None)
    if coco_id is None:
        stats["skipped_no_mapping"] += 1
        print(f"[SKIP:MAP] 매핑 없음: file={src_path.name}, lookup_num={lookup_num}")
        continue

    new_name = f"coco2017_{coco_id}.jpg"
    dst_path = DST_DIR / new_name

    if dst_path.exists() and not OVERWRITE:
        stats["collision"] += 1
        print(f"[SKIP:EXISTS] 이미 존재: {dst_path.name} (원본: {src_path.name})")
        continue

    print(f"[COPY{'-DRY' if DRY_RUN else ''}] {src_path.name}  -->  {new_name}   (lookup_num={lookup_num} -> cocoId={coco_id})")
    if not DRY_RUN:
        # 복사하면서 이름 변경(=목표 이름으로 바로 복사)
        shutil.copy2(src_path, dst_path)
        stats["renamed"] += 1

print("\n==== 요약 ====")
for k, v in stats.items():
    print(f"{k}: {v}")
```

```python
import pandas as pd
import os

# 파일 경로
tsv_path = "/nas/research/03-Neural_decoding/3-bids/2-derivatives/1-beta/beta_huggingface/sub-01/sub-01_mapped_nsd_train.tsv"
csv_path = "/nas/research/03-Neural_decoding/3-bids/2-derivatives/1-beta/beta_nsd/nsd_stim_info_merged.csv"
output_path = "/nas/research/03-Neural_decoding/3-bids/2-derivatives/1-beta/beta_huggingface/sub-01/sub-01_mapped_coco_train.tsv"

# TSV와 CSV 불러오기
tsv_df = pd.read_csv(tsv_path, sep="\t", header=None, names=["filename"])
csv_df = pd.read_csv(csv_path)

# num -> cocoId 매핑 만들기 (num은 0부터 시작한다고 가정)
num_to_coco = dict(zip(csv_df["num"], csv_df["cocoId"]))

def replace_filename(name):
    # 파일명에서 number 추출 (예: coco2017_14.jpg -> 14)
    if name.startswith("coco2017_") and name.endswith(".jpg"):
        number = int(name.replace("coco2017_", "").replace(".jpg", ""))
        # num = number - 1을 cocoId로 변환
        target_num = number - 1
        if target_num in num_to_coco:
            new_number = num_to_coco[target_num]
            return f"coco2017_{new_number}.jpg"
    return name  # 매핑 불가능하면 그대로 반환

# 변환 적용
tsv_df["filename"] = tsv_df["filename"].apply(replace_filename)

# 새로운 파일로 저장
tsv_df.to_csv(output_path, sep="\t", index=False, header=False)

print(f"✅ 변환된 파일이 저장되었습니다: {output_path}")
```

```python
import re, json, numpy as np, pandas as pd, matplotlib.pyplot as plt
from collections import Counter, defaultdict
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

#beta, tsv, coco json 경로

beta_path = "/nas/research/03-Neural_decoding/3-bids/2-derivatives/1-beta/beta_huggingface/sub-01/sub-01_nsdgeneral_train.npy"
tsv_path  = "/nas/research/03-Neural_decoding/3-bids/2-derivatives/1-beta/beta_huggingface/sub-01/sub-01_mapped_coco_train.tsv"
coco_jsons = [
    "/nas/research/03-Neural_decoding/annotations_trainval2017/annotations/instances_train2017.json",
    "/nas/research/03-Neural_decoding/annotations_trainval2017/annotations/instances_val2017.json",
]

# COCO 80개 클래스 → 12개 상위 카테고리 매핑
cat_to_super = {
  1:"person", 2:"vehicle", 3:"vehicle", 4:"vehicle", 5:"vehicle", 6:"vehicle", 7:"vehicle", 8:"vehicle", 9:"vehicle",
  10:"outdoor", 11:"outdoor", 13:"outdoor", 14:"outdoor", 15:"outdoor",
  16:"animal", 17:"animal", 18:"animal", 19:"animal", 20:"animal", 21:"animal", 22:"animal", 23:"animal", 24:"animal", 25:"animal",
  27:"accessory", 28:"accessory", 31:"accessory", 32:"accessory", 33:"accessory",
  34:"sports", 35:"sports", 36:"sports", 37:"sports", 38:"sports", 39:"sports", 40:"sports", 41:"sports", 42:"sports", 43:"sports",
  44:"kitchen", 46:"kitchen", 47:"kitchen", 48:"kitchen", 49:"kitchen", 50:"kitchen", 51:"kitchen",
  52:"food", 53:"food", 54:"food", 55:"food", 56:"food", 57:"food", 58:"food", 59:"food", 60:"food", 61:"food",
  62:"furniture", 63:"furniture", 64:"furniture", 65:"furniture", 67:"furniture", 70:"furniture",
  72:"electronic", 73:"electronic", 74:"electronic", 75:"electronic", 76:"electronic", 77:"electronic",
  78:"appliance", 79:"appliance", 80:"appliance", 81:"appliance", 82:"appliance",
  84:"indoor", 85:"indoor", 86:"indoor", 87:"indoor", 88:"indoor", 89:"indoor", 90:"indoor"
}

# ===== 1) 데이터 로드 =====
B = np.load(beta_path)
names = pd.read_csv(tsv_path, header=None, sep=r"\s+|,|\t", engine="python")[0].astype(str).values
assert len(names) == B.shape[0], "β 행수와 TSV 행수가 다릅니다."

def name_to_imgid(s: str) -> int:
    m = re.search(r'coco2017_(\d+)', s, flags=re.IGNORECASE)
    if not m: raise ValueError(f"image id를 찾을 수 없음: {s}")
    return int(m.group(1))

img_ids = np.array([name_to_imgid(s) for s in names])

# ===== 2) COCO train+val 합치기 =====
imgid_to_cats = defaultdict(list)
cat_meta = {}  # category_id -> {name, super}

for jp in coco_jsons:
    with open(jp, "r") as f:
        coco = json.load(f)
    # category 메타 병합(중복 id는 동일해야 함)
    for c in coco["categories"]:
        cid = c["id"]
        # supercategory는 네가 준 표(cat_to_super)로 통일해서 씀
        cat_meta[cid] = {"name": c["name"], "super": cat_to_super.get(cid, c.get("supercategory", "unknown"))}
    # annotation 병합
    for ann in coco["annotations"]:
        cid = ann["category_id"]
        if cid in cat_to_super:  # 80 클래스 범위만 사용
            imgid_to_cats[ann["image_id"]].append(cid)

# 이미지별 대표 카테고리(최빈값, 동률이면 id 작은 것)
from collections import Counter
def primary_cat_for_image(imgid: int) -> int:
    cats = imgid_to_cats.get(imgid, [])
    if not cats: return -1
    cnt = Counter(cats)
    maxc = max(cnt.values())
    return min([k for k,v in cnt.items() if v == maxc])

primary_cat = np.array([primary_cat_for_image(i) for i in img_ids])
supercats  = np.array([cat_to_super.get(c, "unknown") for c in primary_cat])

# ===== 3) PCA(50) → t-SNE(2D) =====
pca_n = min(50, B.shape[1])
Xp = PCA(n_components=pca_n, random_state=0).fit_transform(B)

perp = max(5, min(30, B.shape[0] // 3))
Z = TSNE(n_components=2, perplexity=perp, learning_rate='auto',
         init='pca', early_exaggeration=12.0, random_state=0).fit_transform(Xp)

# ===== 4) 플롯 유틸 =====
def plot_scatter(Z, labels, title, filename):
    plt.figure(figsize=(7,6))
    uniq = pd.unique(labels)
    if len(uniq) <= 20:
        cmap = plt.cm.get_cmap('tab20', len(uniq))
        for i,u in enumerate(uniq):
            idx = (labels == u)
            plt.scatter(Z[idx,0], Z[idx,1], s=10, alpha=0.85, c=[cmap(i)], label=str(u))
        plt.legend(markerscale=2, frameon=False, bbox_to_anchor=(1.02,1), loc='upper left')
    else:
        if labels.dtype.kind in {'U','S','O'}:
            u2i = {u:i for i,u in enumerate(uniq)}
            lab_int = np.array([u2i[v] for v in labels])
        else:
            lab_int = labels.copy()
        colors = plt.cm.gist_ncar((lab_int - lab_int.min()) / max(1,(lab_int.max()-lab_int.min())))
        if (lab_int == -1).any():
            colors[lab_int == -1] = (0.5,0.5,0.5,0.6)
        plt.scatter(Z[:,0], Z[:,1], s=10, alpha=0.85, c=colors)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(filename, dpi=220)
    plt.close()

# ===== 5) 3가지 색칠 =====
plot_scatter(Z, img_ids,     "t-SNE by image_id (repeats)",     "tsne_by_imgid.png")
plot_scatter(Z, primary_cat, "t-SNE by primary category_id",    "tsne_by_category.png")
plot_scatter(Z, supercats,   "t-SNE by supercategory",          "tsne_by_supercat.png")

print("완료: tsne_by_imgid.png / tsne_by_category.png / tsne_by_supercat.png")
```

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# --- 앞서 만든 변수들 그대로 있다고 가정: B, img_ids, primary_cat, supercats, plot_scatter(), etc. ---

# 1) 앞 30개 인덱스 선택 (데이터 길이보다 길면 자동 조정)
n = min(200, B.shape[0])
idx = np.arange(n)

B90           = B[idx, :]
img_ids90     = img_ids[idx]
primary_cat90 = primary_cat[idx]
supercats90   = supercats[idx]

print("샘플 수:", n)
print("unknown supercategory 개수:", np.sum(supercats90 == "unknown"))

# 2) PCA(<=50, <=n-1) → t-SNE(2D)  (소표본에 안정적인 설정)
pca_n = min(50, B90.shape[1], max(2, n-1))          # n=30이면 최대 29
Xp90  = PCA(n_components=pca_n, random_state=0).fit_transform(B90)

perp = max(5, min(10, (n-1)//3))                    # n=30 → 대략 9~10
Z90   = TSNE(n_components=2, perplexity=perp, learning_rate='auto',
             init='pca', early_exaggeration=12.0, random_state=0).fit_transform(Xp90)

# 3) 플로팅 (파일명에 first30 접미사)
plot_scatter(Z90, img_ids90,     "t-SNE by image_id (first 200)",       "tsne_imgid_first200.png")
plot_scatter(Z90, primary_cat90, "t-SNE by category_id (first 200)",    "tsne_category_first200.png")

print("완료: tsne_*_first200*.png")
```
